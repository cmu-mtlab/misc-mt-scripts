```python
import sys
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
def f(point):
    assert len(point) == 2
    x, y = point
    
    # Goldstein-Price
    return ((1 + (x + y + 1)**2 * (19 - 14*x + 3*x**2 - 14*y + 6*x*y + 3*y**2)) * (30 + (2*x - 3*y)**2 * (18 - 32*x + 12*x**2 + 48*y - 36*x*y + 27*y**2)) / 800000.0) ** (1.0/4.0)

    # Beale's function
    #return (((1.5-x+x*y)**2 + (2.25 - x + x*y**2)**2 + (2.65 - x + x*y**3)**2) / 150000.0) ** (1.0 / 4.0)
    
    #return 0.8 * x * x + 1.0 - np.cos(y * y)
    
def fxy(x, y):
    return f(np.array([x, y]))
```

```python
def dfdd(point, d, epsilon):
    return (f(point + d * epsilon) - f(point - d * epsilon)) / (2 * epsilon)

def dfdx(point, epsilon=1.0e-4):
    d = np.array([1.0, 0.0])
    return dfdd(point, d, epsilon)

def dfdy(point, epsilon=1.0e-4):
    d = np.array([0.0, 1.0])
    return dfdd(point, d, epsilon)

def df(point, epsilon=1.0e-4):
    return np.array([dfdx(point, epsilon), dfdy(point, epsilon)])
```

```python
def sgd_learn(iterations, initial_point, learning_rate=1.0):
    points = []
    x = initial_point.copy()
    points.append(x.copy())
    for i in range(iterations):
        g = df(x)
        dx = -g * learning_rate
        x += dx
        points.append(x.copy())
    return points
```

```python
def adagrad_learn(iterations, initial_point, learning_rate=1.0, epsilon=1.0e-40):
    points = []
    x = initial_point.copy()
    h = np.zeros(2)
    points.append(x.copy())
    for i in range(iterations):
        g = df(x)
        h += g * g
        dx = -g * learning_rate / np.sqrt(h + epsilon)
        x += dx
        points.append(x.copy())
    return points
```

```python
def adadelta_learn(iterations, initial_point, rho=0.95, nepsilon=1.0e-6, depsilon=1.0e-6):
    points = []
    x = initial_point.copy()
    h = np.zeros(2)
    u = np.zeros(2)
    points.append(x.copy())
    for i in range(iterations):
        g = df(x)
        if i > 0:
            h = rho * h + (1.0 - rho) * g * g
            dx = -g * np.sqrt(u + nepsilon) / np.sqrt(h + depsilon)
            u = rho * u + (1.0 - rho) * dx * dx
        else:
            h = g * g
            dx = -g
            u = dx * dx
        x += dx
        points.append(x.copy())
    return points
```

```python
def rprop_learn(iterations, initial_point, learning_rate=1.0, eta_minus=0.5, eta_plus=1.2):
    points = []
    x = initial_point.copy()
    h = np.zeros(2)
    u = np.zeros(2)
    step_sizes = np.ones(2)
    points.append(x.copy())
    for i in range(iterations):
        g = df(x)
        for j in range(len(x)):
            s = g[j] * h[j]
            if s >= 0.0:
                if g[j] > 0.0:
                    dx = -learning_rate * step_sizes[j]
                elif g[j] < 0.0:
                    dx = learning_rate * step_sizes[j]
                h[j] = g[j]
                if s > 0.0:
                    step_sizes[j] *= eta_plus
            else:
                dx = -u[j]
                h[j] = 0.0
                step_sizes[j] *= eta_minus
            step_sizes[j] = max(min(step_sizes[j], 50), 1.0e-6)
            x[j] += dx
            u[j] = dx
        points.append(x.copy())
    return points
```

```python
def quickprop_learn(iterations, initial_point, learning_rate=1.0):
    points = []
    x = initial_point.copy()
    h = np.zeros(2)
    u = np.zeros(2)
    points.append(x.copy())
    for i in range(iterations):
        g = df(x)
        for j in range(len(x)):
            if abs(u[j]) > 1.0e-5:
                a = (g[j] - h[j]) / u[j]
                b = g[j] - a * x[j]
                if a > 0.0:
                    dx = -b / a - x[j]
                    dx = max(dx, -1.0)
                    dx = min(dx, 1.0)
                else:
                    dx = 0.0
            else:
                dx = -g[j] * learning_rate
            x[j] += dx
            u[j] = dx
            h[j] = g[j]
        points.append(x.copy())
    return points
```

```python
def plot_path(points, style='ro'):
    x_vals = [p[0] for p in points]
    y_vals = [p[1] for p in points]
    plt.plot(x_vals, y_vals, style)
```
